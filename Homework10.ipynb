```python
#List all courses and their offerings.Include courses without offerings 
#Hint Use left Join
txt='''
Select *
From Course left join Offering Course.CourseNo = Offering.CourseNo
'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#List all courses who are not offered 
#Hint Use left Join and look for one of column in Course Table whose value is null.
txt='''
SELECT *
FROM Course left join Offering on Course.CourseNo=Offering.CourseNo
WHERE Offering.CourseNo is null
'''
Q2 =  psql.sqldf(txt)
Q2
```

```python
#List all offerings and the faculty assigned to teach them .Also include courses without a faculty assigned to them and Course Name as well 
#Hint:Use Join for 3 tables. Search internet how to do that
txt='''
SELECT *
FROM Course inner join Offering on Course.CourseNo=Offering.CourseNo 
 left join Faculty on Offering.FacNo=Faculty.FacNo
'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#Q4-List courses offfered but not taken any students. 
txt='''
SELECT *
FROM Offering left join Enrollment on Enrollment.Offerno=Offering.Offerno
WHERE Enrollment.Offerno is null
'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#Q5-List  Student LastName and FirstName of the Student with StdGPA greater than average GPA
#Use nested query by calculating average GPA first.
txt='''
SELECT StdFirstName,  StdLastName 
FROM Student
WHERE StdGPA >(select avg(StdGPA) from Student )
'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#Q6-List Faculty information with a maximium and minimum salary
#Hint use nested query to calculate max and min in the form like- where FacSalary=(...) or FacSalary=(....)
txt='''
SELECT *
FROM Faculty
WHERE FacSalary in (select min(FacSalary) from Faculty) or FacSalary in (select max(FacSalary) from Faculty)
'''
Q6 =  psql.sqldf(txt)
Q6
```

```python
#Q7- List faculty name and lastname who are Supervisor of other faculty.
#Use nested faculty by finding Facno of FacSupervisor where FacSupervisor is not null
txt='''
SELECT *
FROM Faculty
WHERE FacSalary>(select avg(FacSalary) from Faculty) or FacSalary in (select min(FacSalary) from Faculty)
'''
Q7 =  psql.sqldf(txt)
Q7
```
